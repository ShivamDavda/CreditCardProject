```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
df = pd.read_csv(r"C:\Users\Shivam\Downloads\New folder (100)\CC GENERAL.csv")
sns.scatterplot(x='BALANCE',y='PURCHASES',data=df,color='blue')
plt.title('Balance vs Purchase')
plt.show()
```

```python
def spendinglevel(purchase):
    if purchase < 500:
        return 'low'
    elif purchase < 2000:
        return 'Medium'
    else: 
        'high'
df['Purchase_level'] = df['PURCHASES'].apply(spendinglevel)
```

```python
print(df['Purchase_level'])
```

```python
df['Payment_flag']=df['PRC_FULL_PAYMENT'].apply(lambda x:'Yes' if x==1 else 'No')
print(df['Payment_flag'])
```

```python
df['Utilized_limit']=df['BALANCE']/df['CREDIT_LIMIT']
df['Utilized_limit']
```

```python
df['Cashadvance_lable']=df['CASH_ADVANCE'].apply(lambda x:'Yes' if x>1 else 'No')
df['Cashadvance_lable']
```

```python
df['Active_puchase']=df['PURCHASES'].apply(lambda x:'Yes' if x>0 else 'No')
df['Active_puchase']
```

```python
df['']
```

```python
df.groupby('Purchase_level')['PURCHASES'].mean().plot(kind='bar',color='skyblue')
plt.title('Purchase vs Purchase level')
plt.show()
```

```python
df.groupby('PRC_FULL_PAYMENT')['CREDIT_LIMIT'].mean().plot(kind='bar',color='lightblue')
plt.title('Credit limit vs full payment')
plt.show()
```

```python
num_df = df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(10,8))
sns.heatmap(num_df.corr(),cmap='coolwarm',annot=True,fmt='.2f')
plt.title('Correlation of numbers')
plt.show()
```
